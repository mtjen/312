```python
# imports
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import classification_report
```

```python
# read the SAS file
full = pd.read_sas('/Users/mtjen/Desktop/312/LLCP2020.XPT')
full.head()
```

```python
# see all the columns in the dataset
cols = full.columns.values.tolist()
cols
```

```python
# remove identifier variables
full.drop(full.iloc[:, 0:32], inplace = True, axis = 1)

altCols = full.columns.values.tolist()
altCols
```

```python
full.shape
```

```python
# move target feature to end of dataframe
targetCol = full.pop('GENHLTH')
full.insert(246, 'GENHLTH', targetCol)
```

```python
# see columns with over 95% of null values
isNullValues = full.isnull().sum()
numSamples = full.shape[0]
threshold = 0.95 * numSamples

for attribute, value in isNullValues.iteritems():
    percentNull = value / numSamples * 100
    if value > threshold:
        print(attribute, value, percentNull)
```

```python
# drop features with all null values 
# most of the other ones with high percentages had reason to have nulls

full.drop(['TOLDCFS', 'HAVECFS', 'WORKCFS'], inplace = True, axis = 1)
full.shape
```

```python
# correlation matrix
corrMatrix = full.corr()
```

```python
# print the attrirbutes with correlation values higher than 0.95
numRows = corrMatrix.shape[0]

colNames = corrMatrix.columns.values.tolist()

for col in corrMatrix.columns:
    for row in corrMatrix.columns:
        value = corrMatrix.at[row, col]
        if value >= 0.99 and row != col:
            print(row, col, value)
```

```python
full.drop(['_TOTINDA', '_DRDXAR2', 'SMOKDAY2', '_AIDTST4', '_CPRACE', 'HTM4',
           '_RFSEAT3', '_MAM5023'], inplace = True, axis = 1)
# adults who excersized past month other than job, any excersize past 30d other than job
# told by doctors they have arthritis, have arthritis
# smoke every/some/no days, every/some/former/no days
# adults HIV tested, had HIV tests
# preferred child race, child non hispanic race including multiracial
# reported height meters, inches
# always/not always wears seatbelt, (almost) always wears
# women 50+ and mammogram in past 2y, women 40+ and mammogram in past 2y

full.shape
```

```python
# need to replace NaN with values to run feature importance algorithm -> Nan to 999
full = full.replace(np.nan, 999)
```

```python
# separate target feature
array = full.values
x = array[:,0:235]
y = array[:,235]

# ensure that Y is the target feature with correct possible values
np.unique(y)
```

```python
# feature extraction using model importance
features = ExtraTreesClassifier()
features.fit(x, y)

importances = features.feature_importances_

print(importances)
```

```python
# get most important 30 feature indices
indices = (-importances).argsort()[:30]
indices
```

```python
for index in indices:
    print(importances[index])
```

```python
# get the column names whose importance is greater than 0.009 to filter
colNames = []

for index in indices:
    if importances[index] > 0.009:
        colNames.append(full.columns[index])
    
print(colNames)
```

```python
full = full.filter(colNames, axis = 1)
full.shape
```

```python
# save cleaned data to csv 
full.to_csv('cleaned.csv', index = False)
```

```python
# import cleaned data 
clean = pd.read_csv('/Users/mtjen/Desktop/312/cleaned.csv')
```

```python
# find how many unique values there are for each feature
clean.nunique()
```

```python
# drop those with lots of values since using categorical naive bayes
clean.drop(['_BMI5', 'WTKG3', '_STSTR', 'WEIGHT2', '_LLCPWT', '_LLCPWT2',
           '_WT2RAKE', '_STRWT'], inplace = True, axis = 1)
```

```python
# separate target feature
cleanArray = clean.values
xClean = cleanArray[:,0:13]
yClean = cleanArray[:,13]
```

```python
# split data to train and test sets
xFullTrain, xFullTest, yFullTrain, yFullTest = train_test_split(x, y, 
                                                                test_size = 0.3, 
                                                                random_state = 42)

xLessTrain, xLessTest, yLessTrain, yLessTest = train_test_split(xClean, yClean, 
                                                                test_size = 0.3, 
                                                                random_state = 42)

print(xFullTrain.shape)
print(xFullTest.shape)
print(yFullTrain.shape)
print(yFullTest.shape)

print()

print(xLessTrain.shape)
print(xLessTest.shape)
print(yLessTrain.shape)
print(yLessTest.shape)
```

```python
# method to get the current time
def getTime():
    now = datetime.now()
    currentTime = now.strftime('%H:%M:%S')
    return currentTime
```

```python
from datetime import datetime
# run algorithms and get begin and end time with the accuracy and classification report
classifier = RandomForestClassifier()

# full dataset
print('Start:', getTime())
classifier.fit(xFullTrain, yFullTrain)
fullPredictions = classifier.predict(xFullTest)
print('End:', getTime())

fullScore = classifier.score(xFullTest, yFullTest)
print('Score:', fullScore)
print(classification_report(yFullTest, fullPredictions))


print()


classifierTwo = RandomForestClassifier()

# less dataset
print('Start:', getTime())
classifierTwo.fit(xLessTrain, yLessTrain)
lessPredictions = classifierTwo.predict(xLessTest)
print('End:', getTime())

lessScore = classifierTwo.score(xLessTest, yLessTest)
print('Score:', lessScore)
print(classification_report(yLessTest, lessPredictions))
```
